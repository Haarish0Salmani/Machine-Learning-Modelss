```python
from sklearn.model_selection import KFold , cross_val_score
```

```python
kf = KFold(n_splits = 5,random_state = 5,shuffle=True)

rm = RandomForestClassifier(random_state = 10, n_jobs = -1)

print(cross_val_score(rm,x,y,cv=kf,scoring="accuracy"))

result = cross_val_score(rm,x,y,cv=kf,scoring="accuracy")

print(result.mean()*100)
```

```python
r_m = RandomForestClassifier(n_estimators=100,random_state = 10,n_jobs=-1)

print(cross_val_score(r_m,x,y,cv = kf,scoring="accuracy"))

result = cross_val_score(r_m,x,y,cv = kf,scoring = "accuracy")

print(result.mean()*100)
```
